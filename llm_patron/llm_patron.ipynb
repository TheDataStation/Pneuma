## Nation Table

```python
nation_schema = "N_NATIONKEY|N_NAME|N_REGIONKEY|N_COMMENT"
```

```python
def create_df(table_name, table_schema):
    df_str = table_schema + "\n" + input_files[table_name]
    df = pd.read_csv(io.StringIO(df_str), sep='|')
    df = df.astype(str)
    return df
```

```python
nation_df = create_df('nation.tbl', nation_schema)
nation_df.head()
```

```python
def query_table(table_df, row_read, query, schema):
    row_count = len(table_df)
    index = 0
    answered = False
    
    final_response = ""
    
    start = time.time()
    while index < row_count:
        input_rows = '\n'.join('|'.join(entry for entry in row) for row in table_df.loc[index:index+row_read].values)
        if (len(input_rows) == 0):
            break
        prompt = f"""I am going to give you a database table. Your task is to analyze the tables
and when given a user query, argue how this table can answer the query.
Here is the schema:\n{schema}
Here is the some rows in the table:\n{input_rows}
Your answer should start with "Yes" if you can answer the query or "No" if you 
cannot answer the query.
Query: {query}"""
        index += row_read

        response = ollama.generate(model="mistral", prompt=prompt)
        if (response['response'].strip()[:2] == "No"):
            final_response = response['response']
            continue
        answered = True
        final_response = response['response']
        break

    end = time.time()
    print(f"Time taken to query table: {end-start}")
    return final_response
```

```python
def query_and_aggregate_answers(table_dfs, row_read, query, schemas, table_names):
    answers = []
    for i in range(len(table_dfs)):
        answers.append(query_table(table_dfs[i], row_read, query, schemas[i]))
        
    final_response = "There is no answer yet."
    for idx, answer in enumerate(answers):
        prompt = f"""I have a collection of database tables and each table is represented by an LLM.
When given a query, each LLM will argue how their table can help answer the query. I have gathered
answers from each LLM, and I now want you to aggregate them all together.
Here is the previously summarized answer:\n{final_response}
Aggregate that with this answer, which comes from the {table_names[idx]} table:\n{answer}"""
        response = ollama.generate(model="mistral", prompt=prompt)['response']
        final_response = response
    return final_response
```

```python
# Read 5 rows at once
print(query_table(nation_df, 5, 'Does China exist in the table?', nation_schema))
```

```python
# Read 25 rows at once
print(query_table(nation_df, 25, 'Does China exist in the table?', nation_schema))
```
